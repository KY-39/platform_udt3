### Задание 1.

$$ y(x) = 1 / (1 + exp(-x)) $$

### Задание 2.

Постройте в одном окне графики функций 
$$ y(x) = sin^2(x)$$ и $$ y(x) = cos^2(x)  $$ на интервале $ [-\pi, +\pi]$
Выведите название осей, название графика и "легенду"
Выведите один график пунктиром и красным цветом.

````
x = np.linspace(-np.pi, np.pi, 100)
plt.plot(x, np.sin(x) ** 2, label='$sin^2(x)$')
plt.plot(x, np.cos(x) ** 2, 'r--', label='$cos^2(x)$')
plt.legend(loc='lower right')
plt.title('Графики функций')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
````

### Задание 3.

Выведите графики функций из задания 2 отдельно, расположив 2 графика горизонтально. Используйте функцию subplot()

````
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Horizontally stacked subplots', y=0)
fig.figsize
ax1.plot(x, np.sin(x) ** 2)
ax2.plot(x, np.cos(x) ** 2)
ax1.set_title('График функции $sin^2(x)$')
ax2.set_title('График функции $cos^2(x)$')
plt.show()
````

### Задание 5.
Постройте диаграмму рассеяния величин $x$ и $y$, зависящих от времени $t$ по закону
$$ x = sin(2t), y =  cos(3t) $$
на интервале $t = [0, 2\pi]$ 

````
t = np.linspace(0, 2 * np.pi, 200)
x = np.sin(2 * t)
y = np.cos(3*t)
plt.scatter(x, y)
plt.show()
````

### Задание 6.
Постройте гистограмму распределения 1000 значений величины, распределенной по нормальному закону с нулевым средним и единичной дисперсией. Укажите число корзин равным 20.

### Задание 7.
Постройте диаграмму рассеяния для 2000 взаимозависимых случайных величин $x$ и $y$, распределенных по нормальному закону с матрицей ковариаций
```

```



````
x = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
plt.scatter(x[:,0], x[:,1])
plt.show()
````
